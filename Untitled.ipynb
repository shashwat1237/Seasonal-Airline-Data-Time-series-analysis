```python
decomp=seasonal_decompose(df,model="additive",period=12)
```

```python
fig,axes=plt.subplots(4,1,figsize=(12,10))
decomp.observed.plot(ax=axes[0],title="Observed")
decomp.trend.plot(ax=axes[1],title="Trend")
decomp.seasonal.plot(ax=axes[2],title="Seasonal")
decomp.resid.plot(ax=axes[3],title="Residuals")
plt.tight_layout()
```

```python
trend_growth_rate = np.polyfit(range(len(decomp.trend.dropna())),decomp.trend.dropna(),1)[0]
print(f"\nTrend Analysis: Approximate annual growth rate = {trend_growth_rate:.1f} thousand passengers/year")
print("Insight: Clear upward trend, indicating expanding air travel demand.")
```

```python
def adf_test(series, title=''):
    result = adfuller(series.dropna())
    print(f'\nADF Test for {title}:')
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    print(f'Critical Values: {result[4]}')
    print('Stationary' if result[1] < 0.05 else 'Non-Stationary')

adf_test(df, 'Original Series')
adf_test(decomp.trend, 'Trend')
adf_test(decomp.resid, 'Residuals')

print("\nStationarity Insight: Original is non-stationary (trend/seasonality); residuals are stationary, suitable for modeling (e.g., ARIMA on residuals).")

# Optional: Multiplicative Decomposition (for comparison, if variance increases)
decomp_mult = seasonal_decompose(df, model='multiplicative', period=12)
fig, axes = plt.subplots(4, 1, figsize=(12, 10))
decomp_mult.observed.plot(ax=axes[0], title='Original (Multiplicative)')
decomp_mult.trend.plot(ax=axes[1], title='Trend (Multiplicative)')
decomp_mult.seasonal.plot(ax=axes[2], title='Seasonality (Multiplicative)')
decomp_mult.resid.plot(ax=axes[3], title='Residuals (Multiplicative)')
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(4,1,figsize=(12,10))
decomp_mult=seasonal_decompose(df,model="multiplicative",period=12)
decomp_mult.observed.plot(ax=axes[0],title="Observed")
decomp_mult.trend.plot(ax=axes[1],title="Trend")
decomp_mult.seasonal.plot(ax=axes[2],title="Seasonal")
decomp_mult.resid.plot(ax=axes[3],title="Residuals")
plt.tight_layout()
```
